```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv("D:\College VIIT\Hackathons\TecKRon\synthetic_dataset_FINAL .csv")
```

```python
# Basic info
print("Dataset Overview:")
print(df.info())
print("\nSummary Statistics:")
print(df.describe())

# Check for missing values
print("\nMissing Values:")
print(df.isnull().sum())
```

```python
# Distribution of Final Prices
plt.figure(figsize=(8,5))
sns.histplot(df['Final Price'], bins=30, kde=True, color='blue')
plt.title("Distribution of Final Prices")
plt.xlabel("Final Price")
plt.ylabel("Frequency")
plt.show()
```

```python
# Select only numeric columns for correlation calculation
numeric_df = df.select_dtypes(include=['number'])

# Plot correlation heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()

```

```python
# Competitor Price Comparisons
plt.figure(figsize=(10,6))
df[['Blinkit Price', 'Zepto Price', 'Instamart Price', 'Final Price']].boxplot()
plt.title("Price Comparisons Across Competitors")
plt.ylabel("Price")
plt.xticks(rotation=45)
plt.show()
```

```python
# Impact of Festive/Seasonal Trends
plt.figure(figsize=(10,5))
sns.boxplot(x='Festive/Seasonal Impact', y='Final Price', data=df)
plt.xticks(rotation=45)
plt.title("Seasonal Impact on Pricing")
plt.show()
```

```python
# Customer Sentiment vs Pricing
plt.figure(figsize=(8,5))
sns.boxplot(x='Customer Sentiment', y='Final Price', data=df)
plt.title("Impact of Customer Sentiment on Pricing")
plt.show()
```

```python
# Time of Ordering Analysis
df['Time of Ordering'] = pd.to_datetime(df['Time of Ordering'])
df['Hour'] = df['Time of Ordering'].dt.hour
plt.figure(figsize=(8,5))
sns.histplot(df['Hour'], bins=24, kde=True, color='green')
plt.title("Order Time Distribution")
plt.xlabel("Hour of the Day")
plt.ylabel("Frequency")
plt.show()
```
